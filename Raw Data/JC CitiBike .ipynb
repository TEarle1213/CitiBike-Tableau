```python
#Import pandas
import pandas as pd
```

```python
#set file path for january 2021
Jan21="2021/JC-202101-citibike-tripdata.csv"
#read it into pandas 
Jan2021=pd.read_csv(Jan21)
#display df
Jan2021.head()
```

```python
#set file path for february 2021
Feb21="2021/JC-202102-citibike-tripdata.csv"
#read it into pandas 
Feb2021=pd.read_csv(Feb21)
#display df
Feb2021.head()
```

```python
#February 2021 is where the data collection and formatting changes, so first I'm going to update the Jan2021 df to be 
#more consistent with the rest of the data. Citibike no longer tracks birth year or gender, bike id was dropped and
#replaced with a transaction id and the formatting for station ids changed. Because we have station locations and names, I'm
#going to drop the station id as well. We'll drop tripduration as well for now, in order to facilitate easier merging.
Jan2021=Jan2021.drop(['birth year', 'gender', 'bikeid', 'start station id', 'end station id', 'tripduration'], axis=1)
#rename the rest of the columns for newer format
Jan2021=Jan2021.rename(columns={'starttime':'started_at','stoptime':'ended_at','start station name':'start_station_name',
                                  'start station latitude':'start_lat','start station longitude':'start_lng','end station name':
                                 'end_station_name','end station latitude':'end_lat','end station longitude':'end_lng'})
Jan2021.info()
```

```python
#From info, we know there are no null values. We do want the Customer/Subscriber values to align with the newer
#Casual/Member format though.
Jan2021['usertype'].replace(to_replace='Subscriber',value='member',inplace=True)
Jan2021['usertype'].replace(to_replace='Customer',value='casual',inplace=True)
Jan2021.head()
```

```python
#convert started and ended times to datetime format
Jan2021['started_at']=pd.to_datetime(Jan2021['started_at'])
Jan2021['ended_at']=pd.to_datetime(Jan2021['ended_at'])
#Check on datatypes
Jan2021.info()
```

```python
#Bring in the rest of the 2021 data
filelist2021=("2021/JC-202103-citibike-tripdata.csv","2021/JC-202104-citibike-tripdata.csv",
             "2021/JC-202105-citibike-tripdata.csv","2021/JC-202106-citibike-tripdata.csv",
             "2021/JC-202107-citibike-tripdata.csv","2021/JC-202108-citibike-tripdata.csv",
              "2021/JC-202109-citibike-tripdata.csv","2021/JC-202110-citibike-tripdata.csv",
             "2021/JC-202111-citibike-tripdata.csv","2021/JC-202112-citibike-tripdata.csv")
# make a list to keep each monthly df, include Feb 
dflist2021=[Feb2021]
#take each file in the list, convert to df, use concatenate to make it a single df
for f in filelist2021:
    df=pd.read_csv(f)
    dflist2021.append(df)   
most2021=pd.concat(dflist2021)
most2021.tail()
```

```python
#format most2021 into the same columns andd datatypes as we updated Jan2021 to.
most2021=most2021.drop(['ride_id', 'rideable_type', 'start_station_id', 'end_station_id'], axis=1)
most2021.info()
```

```python
#cleaning up, change the member_casual column to the easier to read usertype
most2021=most2021.dropna(how='any',axis=0)
most2021=most2021.rename(columns={'member_casual':'usertype'})
```

```python
most2021['started_at']=pd.to_datetime(most2021['started_at'])
most2021['ended_at']=pd.to_datetime(most2021['ended_at'])
most2021.info()
```

```python
JC2021=pd.concat([most2021, Jan2021])
JC2021.head()
```

```python
#and then 2022!
filelist2022=("2022/JC-202201-citibike-tripdata.csv","2022/JC-202202-citibike-tripdata.csv",
             "2022/JC-202203-citibike-tripdata.csv","2022/JC-202204-citibike-tripdata.csv",
             "2022/JC-202205-citibike-tripdata.csv","2022/JC-202206-citibike-tripdata.csv",
             "2022/JC-202207-citbike-tripdata.csv","2022/JC-202208-citibike-tripdata.csv",
             "2022/JC-202209-citibike-tripdata.csv","2022/JC-202210-citibike-tripdata.csv",
             "2022/JC-202211-citibike-tripdata.csv","2022/JC-202212-citibike-tripdata.csv")
# make a list to keep each monthly df, include Feb 
dflist2022=[]
#take each file in the list, convert to df, use concatenate to make it a single df
for f in filelist2022:
    df=pd.read_csv(f)
    dflist2022.append(df)   
JC2022=pd.concat(dflist2022)
JC2022.tail()
```

```python
#clean it up a bit
JC2022=JC2022.drop(['ride_id', 'rideable_type', 'start_station_id', 'end_station_id'], axis=1)
JC2022=JC2022.rename(columns={'member_casual':'usertype'})
JC2022.info()
```

```python
#Drop nulls, fix datatype
JC2022=JC2022.dropna(how='any',axis=0)
JC2022['started_at']=pd.to_datetime(JC2022['started_at'])
JC2022['ended_at']=pd.to_datetime(JC2022['ended_at'])
JC2022.info()
```
