```python
#get pandas
import pandas as pd
```

```python
#set all file paths for 2019
summer19a="201906-citibike-tripdata.csv"
summer19b="201907-citibike-tripdata.csv"
summer19c="201908-citibike-tripdata.csv"
```

```python
#make them into dfs
june2019=pd.read_csv(summer19a)
july2019=pd.read_csv(summer19b)
august2019=pd.read_csv(summer19c)
august2019.head()
```

```python
#One dataframe for Summer of 2019
Summer19=pd.concat([june2019,july2019,august2019])
Summer19.tail()
```

```python
#Checking for nulls
Summer19.count()
```

```python
#Dropping columns that don't appear in more recent data and rows with null values, renaming columns to match current
#format. 
Summer19=Summer19.drop(['birth year','gender','bikeid', 'start station id', 'end station id'], axis=1)
Summer19=Summer19.rename(columns={'starttime':'started_at','stoptime':'ended_at','start station name':'start_station_name',
                                  'start station latitude':'start_lat','start station longitude':'start_lng','end station name':
                                 'end_station_name','end station latitude':'end_lat','end station longitude':'end_lng'})
Summer2019=Summer19.dropna(how='any',axis=0)
Summer2019.count()
```

```python
#Set values to align with more recent data
Summer2019['usertype'].replace(to_replace='Subscriber',value='member',inplace=True)
Summer2019['usertype'].replace(to_replace='Customer',value='casual',inplace=True)
Summer2019.head()
```

```python
#Speed it up for 2020
filelist2020=["202006-citibike-tripdata.csv","202007-citibike-tripdata.csv","202008-citibike-tripdata.csv"]
dflist2020=[]
for f in filelist2020:
    df=pd.read_csv(f)
    dflist2020.append(df)   
Summer2020=pd.concat(dflist2020)
Summer2020.head()
```

```python
#Setting up data for more recent format
Summer2020=Summer2020.drop(['birth year','gender','bikeid','start station id','end station id'], axis=1)
Summer2020=Summer2020.dropna(how='any',axis=0)
Summer2020=Summer2020.rename(columns={'starttime':'started_at','stoptime':'ended_at','start station name':'start_station_name',
                                      'start station latitude':'start_lat','start station longitude':'start_lng','end station name':
                                 'end_station_name','end station latitude':'end_lat','end station longitude':'end_lng'})
Summer2020.count()
```

```python
Summer2020['usertype'].replace(to_replace='Subscriber',value='member',inplace=True)
Summer2020['usertype'].replace(to_replace='Customer',value='casual',inplace=True)
Summer2020.head()
```

```python
#same method as 2020 to make 2021 dataframe
filelist2021=["202106-citibike-tripdata.csv", "202107-citibike-tripdata.csv","202108-citibike-tripdata.csv"]
dflist2021=[]
for f in filelist2021:
    df=pd.read_csv(f)
    dflist2021.append(df)   
Summer2021=pd.concat(dflist2021)
Summer2021.head()
```

```python
#Drop unshared columns, rename usertype, drop nulls 
Summer2021=Summer2021.drop(['rideable_type','ride_id', 'start_station_id','end_station_id'], axis=1)
Summer2021=Summer2021.dropna(how='any',axis=0)
Summer2021=Summer2021.rename(columns={'member_casual':'usertype'})
Summer2021.count()
```

```python
#same method 2022
filelist2022=["202206-citibike-tripdata.csv", "202207-citibike-tripdata.csv","202208-citibike-tripdata.csv"]
dflist2022=[]
for f in filelist2022:
    df=pd.read_csv(f, low_memory=False)
    dflist2022.append(df)   
Summer2022=pd.concat(dflist2022)
Summer2022.head()
```

```python
#cleaning up data on newest frame
Summer2022=Summer2022.drop(['rideable_type','ride_id', 'start_station_id','end_station_id'], axis=1)
Summer2022=Summer2022.dropna(how='any',axis=0)
Summer2022=Summer2022.rename(columns={'member_casual':'usertype'})
Summer2022.count()
```

```python
#Convert start and end times to datetime format
Summer2019['started_at'] = pd.to_datetime(Summer2019['started_at'])
Summer2019['ended_at'] = pd.to_datetime(Summer2019['ended_at'])
Summer2019.dtypes
```

```python
#Apply to the other three dataframes
Summer2020['started_at'] = pd.to_datetime(Summer2020['started_at'])
Summer2020['ended_at'] = pd.to_datetime(Summer2020['ended_at'])
Summer2021['started_at'] = pd.to_datetime(Summer2021['started_at'])
Summer2021['ended_at'] = pd.to_datetime(Summer2021['ended_at'])
Summer2022['started_at'] = pd.to_datetime(Summer2022['started_at'])
Summer2022['ended_at'] = pd.to_datetime(Summer2022['ended_at'])
```

```python
#Add a trip duration column to 2021 and 2022 
Summer2021['tripduration']=Summer2021['ended_at']-Summer2021['started_at']
Summer2022['tripduration']=Summer2022['ended_at']-Summer2022['started_at']
```

```python
#Take the tripduration column in 2019 and 2020 and round it out to the nearest minute
Summer2019['tripduration']=round(Summer2019['tripduration']/60)
Summer2020['tripduration']=round(Summer2020['tripduration']/60)
Summer2019.head()
```

```python
#need datetime to fix timedelta and get minutes for data consistency
import datetime
```

```python
Summer2021['tripduration'] = (Summer2021['tripduration'] / datetime.timedelta(minutes=1)).astype(int)
Summer2022['tripduration'] = (Summer2022['tripduration'] / datetime.timedelta(minutes=1)).astype(int)
Summer2022.head()
```
